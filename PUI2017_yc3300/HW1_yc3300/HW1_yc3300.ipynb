```python
#Generate random distributions to generate 100 2-dimensional random datapoints
```

```python
def ReprRand():
    global ReprRand
    for i in range(1,101):
        np.random.randn(i)
        reprRand = np.random.randn(2,100)
    return ReprRand
```

```python
#Use the same function to generate 50 more random 2*100 numpy arrays
```

```python
import numpy as np
def ReprRandAll():
    global reprRandAll
    reprRandAll = np.zeros((50,2,100))
    for i in range(50):
        np.random.randn(i)
        reprRandAll[i] = 2.5 * np.random.randn(2,100) + 1.5
        i+=1
    return reprRandAll
```
