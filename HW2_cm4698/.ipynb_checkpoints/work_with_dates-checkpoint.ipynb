```python
import os
import sys

import pandas as pd

# Checks that an environmental variable DFDATA exists and that it points to the data facility location 

DFdata = os.getenv("DFDATA")
```

```python
# Uses pandas to read in the CSV file from the DF into a pandas dataframe

df_medallions = pd.read_csv(DFdata + '/h9gi-nx95/1482192000/h9gi-nx95.csv')
df_medallions.head()
```

```python
# Removes all but 2 numerical values columns of my choice
df_medallions = df_medallions[['DATE', 'NUMBER OF PERSONS INJURED']]
df_medallions.head()
```

```python
# setting up some preferences
import pylab as pl
%pylab inline
pl.rc('font', size=15)

#df_medallions['License Expiration Date'] = pd.to_datetime(df_medallions['License Expiration Date'])
df_med = pd.to_datetime(df_medallions['DATE'])

# turning my 2 numerical values columns into x and y data
y = df_medallions['NUMBER OF PERSONS INJURED']
x = df_med

# setting x and y axes labels and title
pl.xlabel("Date")
pl.ylabel("# of people")
pl.title("Number of people injured in collisons by date")
#pl.text(-3, -5000, "scatter plotting one column against another from DFDATA with dates")
pl.xticks(rotation='vertical')

# Plot the columns one against the other in a scatter plot
pl.plot(x, y, ".")
pl.xticks(rotation='vertical')
```
